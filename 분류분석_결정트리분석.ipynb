```python
import pandas as pd
import collections

# feature_name을 불러오거나 정의합니다.
# 예를 들어, feature_name = ['feature1', 'feature2', 'feature2', 'feature3']

# 중복된 항목이 있는지 확인
duplicates = [item for item, count in collections.Counter(feature_name).items() if count > 1]
print(f"중복된 항목: {duplicates}")

# 중복된 항목을 제거하거나 수정
if duplicates:
    feature_name = list(dict.fromkeys(feature_name))
    print("중복된 항목을 제거했습니다. 수정된 feature_name:", feature_name)

# 파일을 불러옵니다.
X_train = pd.read_csv('/Users/jeong-yeji/Desktop/X_train.txt', sep='\s+', names=feature_name, engine='python')
X_test = pd.read_csv('/Users/jeong-yeji/Desktop/X_test.txt', sep='\s+', names=feature_name, engine='python')

Y_train = pd.read_csv('/Users/jeong-yeji/Desktop/Y_train.txt', sep='\s+', header=None, names=['action'], engine='python')
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
# 결정 트리 분류 분석 : 1) 모델 생성
dt_HAR = DecisionTreeClassifier(random_state=156)
```

```python
# 결정 트리 분류 분석 : 2) 모델 훈련
dt_HAR.fit(X_train, Y_train)
```

```python
# 결정 트리 분류 분석 : 3) 평가 데이터에 대한 예측 수행 -> 예측 결과 Y_predict 구하기
Y_predict = dt_HAR.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy = accuracy_score(Y_test, Y_predict)
print('결정 트리 예측 정확도 : {0:.4f}'.format(accuracy))
```

```python
print('결정 트리의 현재 하이퍼 파라미터 : \n', dt_HAR.get_params())
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
params = {
    'max_depth' : [ 6, 8, 10, 12, 16, 20, 24]
}

grid_cv = GridSearchCV(dt_HAR, param_grid=params, scoring='accuracy', 
                       cv=5, return_train_score=True)
grid_cv.fit(X_train , Y_train)
```

```python
cv_results_df = pd.DataFrame(grid_cv.cv_results_)
cv_results_df[['param_max_depth', 'mean_test_score', 'mean_train_score']]
```

```python
print('최고 평균 정확도 : {0:.4f}, 최적 하이퍼 파라미터 :{1}'.format(grid_cv.best_score_ , grid_cv.best_params_))
```

```python
params = {
    'max_depth' : [ 8, 16, 20 ],
    'min_samples_split' : [ 8, 16, 24 ]
}

grid_cv = GridSearchCV(dt_HAR, param_grid=params, scoring='accuracy', 
                       cv=5, return_train_score=True)
grid_cv.fit(X_train , Y_train)
```

```python
cv_results_df = pd.DataFrame(grid_cv.cv_results_)
cv_results_df[['param_max_depth','param_min_samples_split', 'mean_test_score', 'mean_train_score']]
```

```python
print('최고 평균 정확도 : {0:.4f}, 최적 하이퍼 파라미터 :{1}'.format(grid_cv.best_score_ , grid_cv.best_params_))
```

```python
best_dt_HAR = grid_cv.best_estimator_
best_Y_predict = best_dt_HAR.predict(X_test)
best_accuracy = accuracy_score(Y_test, best_Y_predict)

print('best 결정 트리 예측 정확도 : {0:.4f}'.format(best_accuracy))
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
feature_importance_values = best_dt_HAR.feature_importances_
feature_importance_values_s = pd.Series(feature_importance_values, index=X_train.columns)
```

```python
feature_top10 = feature_importance_values_s.sort_values(ascending=False)[:10]
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import export_graphviz
from sklearn.datasets import load_iris

#데이터셋 로드 
data = load_iris()
X, y = data.data, data.target
feature_names = data.feature_names
label_names = data.target_names

# DecisionTreeClassifier 객체 생성 및 학습
best_dt_HAR = DecisionTreeClassifier()
best_dt_HAR.fit(X, y)

# export_graphviz()의 호출 결과로 out_file로 지정된 tree.dot 파일을 생성
export_graphviz(best_dt_HAR, out_file="tree.dot", class_names=label_names, feature_names=feature_names, impurity=True, filled=True)
```
